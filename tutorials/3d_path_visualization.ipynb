# 3D Path Visualization Tutorial

**Let's explore how to visualize 3D paths using the `draw_3d_path` and `draw_3d_dataset` functions from the `cellnav.visualization` module. These functions allow us to create detailed visual representations of 3D paths, which can be particularly useful for analyzing trajectories in 3D space, such as those generated by random walks or other pathfinding algorithms.**

```python
## Dependencies
import numpy as np
from cellnav.visualization import draw_3d_path, draw_3d_dataset
```

## Make some data

```python
## Simulate 3D random walk for 10 particles
num_particles = 10
num_steps = 100

# Generate random walk data
random_walks = np.cumsum(np.random.randn(num_particles, num_steps, 3), axis=1).tolist()
colors = np.tile(np.arange(num_steps) / num_steps, (num_particles, 1))  # Color by step index
```

## Visualize

```python
## Draw single path
path_id:int = 0
path = random_walks[path_id]
color = colors[path_id]

draw_3d_path(path = path,
             color = color, 
             cmap = "viridis",      # colormap for coloring the path
             scalebar = 2,          # length of scalebar in same units as path coordinates
             size = 5,              # radius scale factor for the path
             magnitude = "auto",    # magnitude for scaling the path, can be "auto" or a specific value
             draw_faces = False)    # whether to draw triangular faces for the path segments (if False, will draw lines instead)
```

```python
## Draw the complete dataset of paths
draw_3d_dataset(paths = random_walks,
                colors = colors, 
                cmap = "viridis",      # colormap for coloring the path
                scalebar = 10,          # length of scalebar in same units as path coordinates
                size = 5,              # radius scale factor for the path
                magnitude = "auto",    # magnitude for scaling the path, can be "auto" or a specific value
                draw_faces = False)    # whether to draw triangular faces for the path segments (if False, will draw lines instead)
```
